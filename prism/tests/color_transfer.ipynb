# Color transfer

```python
from prism.scripts.color_transfer import color_transfer
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
src = np.array(Image.open("../images/lion.png")) / 255.
dst = np.array(Image.open("../images/redfish.png")) / 255.

dst_rgb = color_transfer(src, dst)

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(src)
plt.title("Source Image")
plt.axis("off")
plt.subplot(1, 3, 2)
plt.imshow(dst)
plt.title("Destination Image")
plt.axis("off")
plt.subplot(1, 3, 3)
plt.imshow(dst_rgb)
plt.title("Color Transferred Image")
plt.axis("off")
plt.tight_layout()
```
